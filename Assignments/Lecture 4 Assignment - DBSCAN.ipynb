```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Data URL
data_url = "https://archive.ics.uci.edu/static/public/292/data.csv"

# Read the data from the URL
data = pd.read_csv(data_url)

# Features DataFrame (assuming the first columns are features)
X = data.iloc[:, :-1]  # Select all columns except the last (assuming it's the target)

# Scale the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define min_samples based on number of features
min_samples = 2 * X.shape[1]

# Epsilon values for testing
epsilon_values = np.linspace(0.001, 3, 50)

# Lists to store results
percentages_outliers = []

# Loop through epsilon values
for eps in epsilon_values:
  # Perform DBSCAN clustering
  clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(X_scaled)

  # Get cluster labels
  labels = clustering.labels_

  # Count outliers (noise points labeled as -1)
  n_outliers = (labels == -1).sum()

  # Calculate percentage of outliers
  percent_outliers = (n_outliers / len(labels)) * 100

  # Store percentage
  percentages_outliers.append(percent_outliers)

# Line plot of percentage of outliers vs epsilon values
plt.plot(epsilon_values, percentages_outliers)
plt.xlabel("Epsilon Value")
plt.ylabel("Percentage of Outliers (%)")
plt.title("Percentage of Outliers vs Epsilon in DBSCAN")
plt.grid(True)
plt.show()

```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Data URL
data_url = "https://archive.ics.uci.edu/static/public/292/data.csv"

# Read the data from the URL
data = pd.read_csv(data_url)

# Features DataFrame (assuming the first columns are features)
X = data.iloc[:, :-1]  # Select all columns except the last (assuming it's the target)
feature_names = X.columns  # Store feature names for later reference

# Scale the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define min_samples based on number of features
min_samples = 2 * X.shape[1]

# Reasonable Epsilon Value (choose based on plot analysis)
chosen_epsilon = 2  # Replace with your chosen value based on the plot analysis (from previous step)

# Perform DBSCAN clustering
clustering = DBSCAN(eps=chosen_epsilon, min_samples=min_samples).fit(X_scaled)

# Get cluster labels
labels = clustering.labels_

# Separate data points based on labels (clusters and outliers)
data_clustered = data[labels != -1]  # Data points in clusters
data_outliers = data[labels == -1]  # Outliers

# Function to calculate and display mean spending per category
def compare_means(data, title):
  print(f"\n** {title} Mean Spending per Category:")
  for feature in feature_names:
    mean_value = data[feature].mean()
    print(f"  - {feature}: {mean_value:.2f}")

# Compare means for clustered data
compare_means(data_clustered, "Clustered Data")

# Compare means for outliers (if any)
if len(data_outliers) > 0:
  compare_means(data_outliers, "Outlier Data")
else:
  print("\n** Outlier Data: No outliers identified by DBSCAN.")

```
