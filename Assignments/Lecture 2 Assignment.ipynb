```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
hepatitis_data = hepatitis.data
hepatitis_target = hepatitis.target

# Convert data to DataFrame
data = pd.DataFrame(data=hepatitis_data, columns=hepatitis.variables['name'])
data['Class'] = hepatitis_target
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'data_url' with the actual URL of the dataset)
data_url = 'https://archive.ics.uci.edu/static/public/46/data.csv'
data = pd.read_csv(data_url)

# Mapping for Class column
class_mapping = {1: 'DIE', 2: 'LIVE'}
data['Class'] = data['Class'].map(class_mapping)

# Chapter 1: Demographic Dynamics
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(data['Age'], bins=8, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')

plt.subplot(1, 2, 2)
sns.countplot(x='Sex', data=data, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Chapter 2: Symptoms Spectrum
symptoms = ['Fatigue', 'Malaise', 'Anorexia']
plt.figure(figsize=(12, 5))
for i, symptom in enumerate(symptoms, 1):
    plt.subplot(1, len(symptoms), i)
    sns.countplot(x=symptom, data=data, palette='pastel')
    plt.title(f'{symptom} Distribution')
    plt.xlabel(symptom)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Chapter 3: Treatment Trajectories
treatments = ['Steroid', 'Antivirals']
plt.figure(figsize=(12, 5))
for i, treatment in enumerate(treatments, 1):
    plt.subplot(1, len(treatments), i)
    sns.countplot(x=treatment, data=data, palette='pastel')
    plt.title(f'{treatment} Distribution')
    plt.xlabel(treatment)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Chapter 4: Clinical Correlations
clinical_tests = ['Bilirubin', 'Alk Phosphate', 'Sgot', 'Albumin']
plt.figure(figsize=(12, 10))
for i, test in enumerate(clinical_tests, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data[test], bins=20, kde=True, color='skyblue')
    plt.title(f'{test} Distribution')
    plt.xlabel(test)
plt.tight_layout()
plt.show()

# Additional: Class Distribution
plt.figure(figsize=(8, 6))
data['Class'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Patient Classes')
plt.axis('equal')
plt.show()
```
